###**MENGATASI MISING VALUES**

Menghapus mising values yang presentasenya lebih dari 70%

```python
df.drop(['Instructions', 'Rating',
         'Review', 'Cancellation / Rejection reason', 'Restaurant compensation (Cancellation)',
         'Restaurant penalty (Rejection)', 'Customer complaint tag'], axis=1, inplace=True)
```

Melakukan imputasi pada mising values yang presentasinya 1,5%

```python
df['KPT duration (minutes)'].dropna().describe()
```

```python
df['KPT duration (minutes)'] = df['KPT duration (minutes)'].fillna(df['KPT duration (minutes)'].dropna().mean())
```

##**Tahap 4 : melakukan pengecekan outliers**

```python
print(df.describe())
```

Menampilkan outliers di setiap kolom

```python
for col in df.select_dtypes(include=['number']).columns:
    mean = df[col].mean()
    std = df[col].std()
    lower_bound = mean - (3 * std)
    upper_bound = mean + (3 * std)

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    print(f"Outliers di kolom {col}: {len(outliers)} baris")
```
